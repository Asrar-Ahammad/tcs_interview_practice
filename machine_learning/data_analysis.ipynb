## Normal Data Distribution
- In probability theory this kind of data distribution is known as the normal data distribution, or the Gaussian data distribution, after the mathematician Carl Friedrich Gauss who came up with the formula of this data distribution.

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.normal(5.0, 1.0, 10000)
plt.hist(x, 100)
plt.show()
```

```python
"""
This code generates random data points, performs a linear regression on them, and plots the resulting line.

The `x` and `y` variables are created by generating 100 random data points from normal distributions with different means and standard deviations.

The `stats.linregress()` function is used to perform the linear regression, returning the slope, intercept, correlation coefficient, p-value, and standard error of the regression.

The `myfunc()` function is defined to calculate the y-value for a given x-value using the slope and intercept from the linear regression.

The `mymodel` list is created by applying the `myfunc()` function to each x-value in the `x` list, generating the y-values for the regression line.

Finally, the code creates a scatter plot of the original data points and overlays the regression line using `plt.scatter()` and `plt.plot()`.
"""
import matplotlib.pyplot as plt
from scipy import stats

x = numpy.random.normal(1.0, 1.0, 100)
y = numpy.random.normal(1.0, 5.0, 100)

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
